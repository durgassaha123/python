#**Raising an Exception**

In Python, we use **raise** to throw an exception if a condition occurs. We can complement the statement with a custom exception as well.

image.png

**Example**

```
number = int(input('Enter a number'))
if number < 10:
  raise Exception('Sorry, we only accept numbers greater 10, you entered {}'.format(number))
```

##The **Try** and **Except** Block

When an exception occurs in Python, it will normally stop running the code and throws an exception.
We can handle these exceptions using the **try** and **except** statement.

image.png

**Example**

```
m= 10

print(m)
```

```
print(w)
```

```
try:
  print(w)
except:
  print("I think you have not defined 'W'")
```

```
number1 = int(input('Enter first number '))
number2 = int(input('Enter second number '))

try:
  answer = number1/number2
  print("That's great, we are able to perform division. Answer is : ", answer)
except:
  print('Sorry, we are unable to perform division, try different numbers')
```

```
try:
    a=5
    b='0'
    print(a/b)
except:
    print('Some error occurred.')
print("Out of try except blocks.")
```

#**The else Clause**

We can use the **else** statement in Python to execute a different block of code if there are **No** exceptions raised.

image.png

```
try:
    c = 90/30
    print('Answer: ', c)
except:
    print('Why are you trying to divide by 0?')
else:
    print('Executing the else clause because there was no exceptions.')
```

<br>

#***finally*** Cluase

This ***finally*** clause is executed regardless if there are exceptions or not. It helps to release external resources after excuting your code.

image.png

```
try:
    c = 90/0
    print('Answer: ', c)
except:
    print('Why are you trying to divide by 0?')
else:
    print('Executing the else clause because there was no exceptions.')
finally:
  print('This statement is excuted whether there is an exception or not')
```

<br>


#**Summary**

* **raise** allows you to throw an exception at any time.
* **assert** enables you to verify if a certain condition is met and throw an exception if it is not.
* in the **try** clause, all statements are executed until an exception is encountered.
* **except** is used to catch and handle the exception(s) that are encountered in the try clause.
* **else** lets you code sections that should run only when no exceptions are encountered in the try clause.
* **finally** enables you to execute sections of code that should always run, with or without any previously encountered exceptions.

#**Python Syntax Errors**

In Python, **Syntax Error** happens when you don't follow the proper syntax or structure of writing Python code.

For example: missing colon, missing brackets, missing quotes, wrong spellings etc.

```
# let's say we want to add two numbers
a = 10
b = 20
c = a + b

print(c)
```

```
# now if we don't follow the correct syntax, we will get a syntax error

a = 10
b = 20
c = a +

print(c)
```

From above, we can see that we have syntax error because we forgot to bring the ***b*** variable. This error will be solved if we follow proper syntax by adding the ***b*** variable.

```
# let's say we want to define a function
#to perform to perform multiplication of two numbers

def mult(a,b)
results = a*b
return results
```

We have syntax error in the above code because we ommitted a **colon(**:**).**

let's add the colon and see how it will work

```
def mult(a,b):
  results = a*b
  return results
```

```
mult(6,10)
```

*Now we see that everything works very fine*.

There are three common ways that you can mistakenly use keywords:

* **Misspelling** a keyword
* **Missing** a keyword

**Misspelling** a keyword

```
fro i in range(5):
  print(i)
```

We can see from the above that we misspelled the Python keyword ***for*** and because of that we are getting a ***SyntaxError: invalid syntax***

```
for i in range(5):
  print(i)
```

All good now

```
colors= ['Green','Red','Brown','Black','Blue','White']
if 'Red' colors:
  print('Red is in colors')
```

#**Runtime Errors**

The second type of error is a ***runtime error***. A program with a runtime error is one that passed the interpreters syntax checks, and started to execute. However, during the execution of one of the statements in the program, an error occurred that caused the interpreter to stop executing the program and display an error message. Runtime errors are also called exceptions because they usually indicate that something exceptional (and bad) has happened.

*Here are some examples of common runtime errors you are sure to encounter:*

* Misspelled or incorrectly capitalized variable and function names (Use an undefined variable or function)

* Attempts to perform operations (such as math operations) on data of the wrong type (ex. attempting to subtract two variables that hold string values)

* Dividing by zero

* Attempts to use a type conversion function such as int on a value that cant be converted to an int

```
Price = float(input('Enter item price: '))

if Price > 200:
  print("This is too costly, I can't buy")
else:
  print("This is ok for me to buy")
```
